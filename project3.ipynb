```python
import torch
import cv2

# تحميل موديل YOLOv5 المدرب مسبقًا (مديل yolov5s)
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)

# فتح الفيديو (يمكنك تغير 'video.mp4' لمصدر الفيديو لديك)
cap = cv2.VideoCapture('projects/854141-hd_1920_1080_25fps.mp4')

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    # كشف الأشياء على الفريم الحالي
    results = model(frame)

    # رسم النتائج على الصورة
    annotated_frame = results.render()[0]

    # عرض الفريم مع التحديدات
    cv2.imshow('YOLOv5 Detection', annotated_frame)

    # اضغط 'q' للخروج
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# تحميل موديل YOLOv5s الجاهز (محتاج إنترنت أول مرة)
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)

cap = cv2.VideoCapture('')

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # كشف الأشياء في الإطار
    results = model(frame)

    # رسم مربعات الكشف على الإطار
    annotated_frame = results.render()[0]

    # عرض الفيديو مع الكشف
    cv2.imshow('YOLOv5 Object Detection', annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
